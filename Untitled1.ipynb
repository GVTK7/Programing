```python
%%writefile pairs.cpp

#include<iostream>
using namespace std;

int main()
{
    int i,j,n;
     
     n=4;
 
     for(i=1;i<=n-1;i++)
     {
         for(j=i+1;j<=n;j++)
         {
             cout<<i<<" "<<j<<endl;
         }
     }
}
```

```python
#include <math.h>
#include <iostream>
using namespace std;

int main()
{
    int i,n,j;
    cin>>n;
    for (i=1;i<=2*n-1;i++)
    {
        for (j=1;j=n-abs(i-n);j++)
        {
            cout<<"* ";
        }
        cout<<endl;
    }
    return 0;
}
```
