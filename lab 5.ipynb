```python
//Given an array A of n integers, write a program to find the number of valid pairs in A. A pair (A[i], A[j]) is said to be a valid pair, if i < j and A[i] < A[j].
%%writefile p2.cpp
#include<iostream>
#include<stdlib.h>
using namespace std;


int main()
{
    int i,j,k,n,s;
    cin>>n;
    int a[n];
    s=0;
 
    for(i=0;i<=n-1;i++)
      cin>>a[i];
 
    for(j=0;j<=n-1;j++)
    {
        for(k=j+1;k<=n-1;k++)
        {
            if(a[j]<a[k])
            {
                s=s+1;
            }
        }
    }
    cout<<s;
}
```

```python
//Given an array A of n integers, write a program to move all zeros in A to the end of the array. The relative order of the non-zero elements in the array should not be disturbed. For the input A ={21,0,4,0,55,0,6}, the output is {21,4,55,6,0,0,0}
%%writefile p3.cpp
#include<iostream>
#include<stdlib.h>
using namespace std;

int main()
{
    int i,j,k,n,c,x;
    cin>>n;
    int a[n],b[n];

    for(i=0;i<=n-1;i++)
    {  
        cin>>a[i];
        b[i]=a[i];
    }
    c=0;
 
    for(j=0;j<=n-1;j++)
    {
        if(a[j]==0)
        {
            for(k=j;k<=n-2;k++)
            {
                b[k-c]=b[k-c+1];
                b[k-c+1]=0;
            }
            c=c+1;
        }
    }
    
    for(x=0;x<=n-1;x++)
      cout<<b[x]<<" ";
 
    return 0;
}
```
